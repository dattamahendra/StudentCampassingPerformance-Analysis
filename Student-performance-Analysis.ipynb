```python
# Assigning grades to the grades according to the following criteria :
# 0  - 40 marks : grade E
# 41 - 60 marks : grade D
# 60 - 70 marks : grade C
# 70 - 80 marks : grade B
# 80 - 90 marks : grade A
# 90 - 100 marks : grade O

def getgrade(percentage, status):
  if status == 'Fail':
    return 'E'
  if(percentage >= 90):
    return 'O'
  if(percentage >= 80):
    return 'A'
  if(percentage >= 70):
    return 'B'
  if(percentage >= 60):
    return 'C'
  if(percentage >= 40):
    return 'D'
  else :
    return 'E'

data['grades'] = data.apply(lambda x: getgrade(x['percentage'], x['status']), axis = 1 )

data['grades'].value_counts()
```

```python
dataconclusion = pd.read_csv('student_dataset.csv')
print(dataconclusion.shape)
dataconclusion.head()
```

```python
def makeConclusion(k,tech,comm,team,ana):
    s=''
    if k=='not success':
        return 'all type of updation training required'
    elif k=='Lavel A':
        return "No updation required"+str(type(tech))+str(comm)+str(team)+str(ana)
    elif k=='Lavel B':
        #return 'upgradation for A'+str(type(tech))+str(comm)+str(team)+str(ana)
        if tech<7:
            s+="tech,"
        if comm<7 :
            s+= "Communication, "
        if  team<7 :
            s+= "Teamwork, "
        if  ana<7:
            s+= "Analytical Skill "
        if tech<7 and comm<7 and team<7 and ana<7:
            return "All type upgradation required"
        else:
            return s+' upgradation training required for getting chance in  A lavel company'
        
    elif k=='Lavel C':
        return 'upgradation for B'+str(tech)+str(comm)+str(team)+str(ana)
        if tech<5:
            s+="tech,"
        if comm<5:
            s+= "Communication, "
        if team<5 :
            s+= "Teamwork, "
        if ana<5:
            return "Analytical Skill upgradation required for getting chance in Company Lavel B"
        if tech<5 and comm<5 and team<5 and ana<5:
            return "All type upgradation required"
        else:
            return s+' upgradation training required for getting chance in  A lavel company'
        
dataconclusion['conclusion'] = dataconclusion.apply(lambda x: makeConclusion(x['Company Lavel'],x['tech'],
                            x['communicating'],x['teamwork'],x['analysis']), axis = 1 )
```
